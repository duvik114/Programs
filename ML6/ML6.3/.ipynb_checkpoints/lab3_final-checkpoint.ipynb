```python
import re
import keras
import numpy as np

from keras.callbacks import ModelCheckpoint
from keras.layers import Dense, LSTM
from collections import defaultdict
```

```python
text = open('text.txt').read().lower()
text = re.sub(r'[ё]', 'е', text)
text = re.sub(r"[^а-я\.\s+]+", "", text)
text = re.sub(r'\s+', ' ', text)
text = re.sub(r'(\.+\s*)+', '. ', text)
text = re.sub(r'\s*\.+\s*', '. ', text)
print(text)
```

```python
text_chars = sorted(list(set(text)))
chars = dict((c, i) for i, c in enumerate(text_chars))
indexes = dict((i, c) for i, c in enumerate(text_chars))
print(chars)
print(indexes)
```

```python
seq_len = 12
dataX = list(s[i:i + seq_len] for s in text.split('.') for i in range(len(s) - seq_len))
dataY = list(s[seq_len + i] for s in text.split('.') for i in range(len(s) - seq_len))
print(dataX[:10])
print(dataY[:10])
```

```python
X = np.zeros((len(dataX), seq_len, len(chars)))
Y = np.zeros((len(dataX), len(chars)))
for i, s in enumerate(dataX):
    for j, ch in enumerate(s):
        X[i, j, chars[ch]] = 1
    Y[i, chars[dataY[i]]] = 1

print(dataX[242], np.argmax(X[242]), indexes[np.argmax(X[242])])
print('One_hot:', X[242][0], '\nLength:', len(X[242][0]))
print('dataX shape', X.shape)
print('dataY shape', Y.shape)
```

```python
# filepath="weights/improvement-{epoch:02d}-{loss:.4f}-bigger.hdf5"
# checkpoint = ModelCheckpoint(filepath, monitor='loss', verbose=1, save_best_only=True, mode='min')
# callbacks_list = [checkpoint]
model = keras.Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(Y.shape[1], activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy')
model.fit(X, Y, epochs=30)
# model.fit(X, Y, epochs=30, callbacks=callbacks_list)
model.summary()
```

```python
def get_one_hot(cur_sentence):
    cur_one_hot = np.zeros((1, seq_len, len(text_chars)))
    for i, ch in enumerate(cur_sentence):
        cur_one_hot[0, i, chars[ch]] = 1
    return cur_one_hot


for _ in range(4):
    cur_sentence = dataX[np.random.randint(len(dataX) - 1)]
    sentence = cur_sentence
    res = cur_sentence
    for _ in range(40):
        cur_one_hot = get_one_hot(cur_sentence)
        predict = model.predict(cur_one_hot, verbose=0)
        next_ch = indexes[np.argmax(predict)]
        cur_sentence += next_ch
        cur_sentence = cur_sentence[1:]
        res += next_ch
    print('Sentence =', sentence)
    print('Predict =', res, '\n')
```

```python
continues = defaultdict(lambda: defaultdict(lambda: 0))
for s, c in zip(dataX, dataY):
    continues[s][c] += 1
for sent in dataX:
    vals_sum = sum([v for v in continues[sent].values()])
    for cont in continues[sent]:
        continues[sent][cont] /= vals_sum
```

```python
for _ in range(4):
    cur_sentence = dataX[np.random.randint(len(dataX) - 1)]
    sentence = cur_sentence
    res = cur_sentence
    for _ in range(40):
        nexts = continues[cur_sentence]
        p = list(nexts.values())
        if len(p) == 0:
            break
        next_ch = list(nexts.items())[np.random.choice(len(p), p=np.array(p))][0]
        cur_sentence += next_ch
        cur_sentence = cur_sentence[1:]
        res += next_ch
    print('Sentence =', sentence)
    print('Predict =', res, '\n')
```
